```python
dateiZugriff = open("1.txt", "r")
inhalt = dateiZugriff.read()
zeilen = inhalt.splitlines()

print(zeilen)
```

```python
linkeListe = []
rechteListe = []

for zeile in zeilen:
    links, rechts = zeile.split("   ")
    linkeListe.append(int(links))
    rechteListe.append(int(rechts))

linkeListe.sort()
rechteListe.sort()

print(linkeListe)
print(rechteListe)
```

```python
entfernungGesamt = 0

for zähler in range(len(linkeListe)):
    entfernung = linkeListe[zähler] - rechteListe[zähler]
    entfernung = abs(entfernung)
    entfernungGesamt = entfernungGesamt + entfernung

print(entfernungGesamt)
```
