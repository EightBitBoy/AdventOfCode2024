```python
input = open("3.txt", "r")
text = input.read()
print(text)
```

```python
lines = text.split("mul(")
print("\n".join(lines))
```

```python
import re

regex = "^\d{1,3},\d{1,3}\)"
matches = []
for line in lines:
    match = re.search(regex, line)
    if match != None:
        matches.append(match.group())

print(matches)
```

```python
sum = 0
for item in matches:
    pair = item.replace(")", "")
    left, right = pair.split(",")
    sum += int(left) * int(right)

print(sum)
```
